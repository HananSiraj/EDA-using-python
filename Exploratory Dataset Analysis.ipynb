# 1.How many are there in each team and the percentage splitting with respect to the total employees.

```python
each_team=df['Team'].value_counts()
print("Number of employees in each team:")
each_team
```

```python
team_percent=round((each_team/len(df))*100,2)
print("Percentage splitting with respect to the total employees:")
team_percent
```

```python
plt.figure(figsize=(8, 8))
plt.pie(team_percent, labels=team_percent.index, autopct='%1.1f%%',startangle=360,colors=['salmon','skyblue','lightgreen','orange','olive'])
plt.title('Percentage Splitting of Employees in Each Team')
plt.show()
```

##### Here we created a pie chart based on the exploratoryanalysis of percentage splitting of employees in each team.here we can see the team 'New Orleans Pelican' have the highest rate of employees and the team 'Orlando Magic' & 'Minnesota Timberwolves' got the least percentage of employees

## 4.Find out under which team and position, spending in terms of salary is high.

```python
team_position_salary_sum = df.groupby(['Team', 'Position'])['Salary'].sum()
max_spending_team_position = team_position_salary_sum.idxmax()
max_spending_value = team_position_salary_sum.max()
print("Team and Position with the Highest Salary Spending:")
print("Team:", max_spending_team_position[0])
print("Position:", max_spending_team_position[1])
print("Total Salary Spending:", max_spending_value)

```

```python
plt.figure(figsize=(6,3))
sns.boxplot(x='Salary', data=df)
plt.title('Boxplot of Salary')
plt.show()
```

##### Here boxplot visualizes the distribution of the 'Salary' column, A boxplot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset and provides a summary of the central tendency, spread, and potential outliers within the data. 


##### 1.Box (Interquartile Range - IQR): The box represents the interquartile range, which is the middle 50% of the data. The lower (Q1) and upper (Q3) edges of the box indicate the 25th and 75th percentiles of the data, respectively.


##### 2.Median (Q2): A line inside the box represents the median, which is the middle value of the dataset. It divides the data into two equal halves. 

##### 3.Whiskers: The whiskers extend from the edges of the box to the minimum and maximum values within a specified range. 

#####  4.Outliers: Points beyond the whiskers are considered outliers.this also indicate extreme values in the dataset that are significantly different from the rest of the data. Here we created a simple boxplot of the 'Salary' column, providing an overview of the distribution, median and potential outliers.

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='Team', y='Salary', hue='Position', data=df)
plt.title('Boxplot of Salary by Team and Position',color='r')
plt.xticks(rotation='vertical')
plt.show()
```

##### Here created a more complex boxplot. It shows the distribution of 'Salary' within different 'Teams' and distinguishes 'Positions' within each team. This can help in understanding how salaries vary not only across teams but also within different positions within each team.These visualizations are valuable for quickly understanding the key characteristics of the salary distribution and its variations across different teams and positions.

## 5.Find if there is any correlation between age and salary , represent it visually.

```python
corr=df[['Age','Salary']].corr()
print("Correlation between age and salary:")
print(corr)
```

#### Heat map to represent correlation visually

```python
sns.heatmap(corr,annot=True,cmap="viridis", fmt=".2f")
plt.title('Correlation between Age and Salary',color='blue')
plt.show()
```

##### Here we created Heatmap that shows the Correlation between'Age' and'Salary',By using heatmap we colud able to plot the data that shows the magnitude of phenomenon in the form of colors.A positive correlation indicates that as one variable increases, the other tends to increase, while a negative correlation indicates an inverse relationship.Here 'Age vs Age ' and 'Salary vs Salary' have positive correlation 1 whereas 'Age vs 'Salary' and 'Salary' vs 'Age' have 0.21 positive correlation but relatively weak.Thus we can conclude that as the employees age increases there is a slight probablitity to increase the Salary
